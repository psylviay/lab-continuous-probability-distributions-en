```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Function to generate uniformly distributed random numbers
def generate_uniform_random_numbers(bottom, ceiling, count):
    # Generate uniform random numbers between bottom and ceiling
    return stats.uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)

# Call the function with the first set of parameters
random_numbers_1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)

# Call the function with the second set of parameters
random_numbers_2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)
```

```python
# Plot the uniform distributions as histograms
plt.figure(figsize=(12, 6))

# Plot for the first set of parameters
plt.subplot(1, 2, 1)
plt.hist(random_numbers_1, bins=10, edgecolor='black', color='skyblue')
plt.title('Uniform Distribution (10 to 15)')
plt.xlabel('Value')
plt.ylabel('Count')

# Plot for the second set of parameters
plt.subplot(1, 2, 2)
plt.hist(random_numbers_2, bins=10, edgecolor='black', color='lightcoral')
plt.title('Uniform Distribution (10 to 60)')
plt.xlabel('Value')
plt.ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

# Differences Between the Two Distributions

The two uniform distributions in the challenge differ in several ways:

### 1. **Range (Spread) of Values**
   - **First Distribution (10 to 15)**: The distribution spans a **range of 5 units** (from 10 to 15).
   - **Second Distribution (10 to 60)**: The distribution spans a **range of 50 units** (from 10 to 60).

 This wider range in the second distribution results in the values being more spread out compared to the first distribution. As a result, the histogram for the second distribution will likely show a larger **spread** of values, with data points distributed over a broader x-axis.

### 2. **Number of Data Points (Sample Size)**
   - **First Distribution**: Generates **100 random numbers** between 10 and 15.
   - **Second Distribution**: Generates **1,000 random numbers** between 10 and 60.

The second distribution has **10 times more data points**, which means its histogram will have more **bars filled** (with higher frequency counts) compared to the first one. This larger sample size gives a more accurate representation of the uniform distribution.

### 3. **Shape of the Histograms**
   - **First Distribution**: With fewer data points (100) and a smaller range (5), the histogram will be more **compact** and will cover only the narrow range of 10 to 15.
   - **Second Distribution**: With more data points (1,000) and a larger range (50), the histogram will be **smoother** and show a **more spread-out distribution** across the wider range of 10 to 60.

### 4. **Granularity**
   - **First Distribution**: Due to fewer data points (100) spread over 5 units, the histogram will have **coarser granularity**. Each bin will likely contain more data points, and there will be fewer visible changes between the bars.
   - **Second Distribution**: With 1,000 data points and a range of 50 units, the histogram will have **finer granularity**, showing more detailed changes in frequency between bins.

# Exponential Distribution

The following code generates **exponentially distributed random numbers** using `numpy.random.exponential`. The exponential distribution is commonly used to model the time between events in a Poisson process.

### Key Parameters:
- The **mean** of the distribution is **10**.
- Two sequences of exponentially distributed numbers will be generated:
  - One with a size of 10.
  - One with a size of 100.

The distributions will then be plotted using histograms with 100 bins.

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Generates exponentially distributed numbers
def generate_exponential_numbers(mean, size):
    return np.random.exponential(scale=mean, size=size)

# Generates the first sequence of size 10 with mean 10
data_1 = generate_exponential_numbers(mean=10, size=10)

# Generates the second sequence of size 100 with mean 10
data_2 = generate_exponential_numbers(mean=10, size=100)

# Plots the distributions
plt.figure(figsize=(12, 6))

# Plot for the first sequence (size=10)
plt.subplot(1, 2, 1)
plt.hist(data_1, bins=100, edgecolor='black', color='skyblue')
plt.title('Exponential Distribution (size=10, mean=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot for the second sequence (size=100)
plt.subplot(1, 2, 2)
plt.hist(data_2, bins=100, edgecolor='black', color='lightcoral')
plt.title('Exponential Distribution (size=100, mean=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

## Ways in which the distributions are different

**Mean** - their means are different, so the distribution changes between both plots presented.

**Sample Size** - the first data set has only 10 data points whereas the second one has 100.
This difference makes the first distribution appear more categorical, not to mention that
its shape is more difficult to evaluate versus the second one that is smoother.

**Shape of the Histogram** - because of the small number of data points, the first distribution does

**Granularity and Accuracy** - the first distribution has lower granularity as the data points do not fill the bins.
The second distribution has higher granularity and thus displays more details around the distribution.

In short, the distribution with the 100 data points provides a smoother, more accurate representation of the 
**exponential distribution**.

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

# Exponential Distribution Probability

In this problem, the amount of time a customer spends in the bank is **exponentially distributed** with a mean of **10 minutes**. The **rate parameter** \(\lambda\) is the inverse of the mean:

\[
\lambda = \frac{1}{10} \, \text{(since the mean is 10 minutes)}.
\]

The **exponential distribution** is given by the cumulative distribution function (CDF):

\[
P(X \leq x) = 1 - e^{-\lambda x}
\]

Where:
- \(X\) is the random variable representing the time spent in the bank,
- \(\lambda = \frac{1}{10}\) is the rate parameter,
- \(x = 15\) is the value we are interested in (15 minutes).

We want to calculate the probability that a customer spends **less than 15 minutes** in the bank:

\[
P(X < 15) = 1 - e^{-\frac{1}{10} \times 15}
\]

---

```python
import numpy as np

# Given data
lambda_ = 1/10  # Rate parameter (1/mean)
x = 15          # Time in minutes

# Calculate the cumulative distribution function (CDF) at x = 15
probability = 1 - np.exp(-lambda_ * x)

# Print the probability
print(f"The probability that a customer spends less than 15 minutes in the bank is: {probability:.4f}")
```
